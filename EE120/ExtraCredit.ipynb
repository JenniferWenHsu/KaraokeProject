```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile as wav
# data = wav.read('original2.wav')[1]
# plt.plot(data)
# plt.show()
```

```python
fs, data = wav.read('original2.wav') # load the data
a = data.T[0] # this is a two channel soundtrack, I get the first track
b=[(ele/2**8.)*2-1 for ele in a] # this is 8-bit track, b is now normalized on [-1,1)
test = np.fft.fft(b) # calculate fourier transform (complex numbers list)


maxAmplitude = np.absolute(np.max(test))

#modifiedOriginal = [elem/maxAmplitude for elem in test]
modifiedOriginal = test
d = int(len(test)/2)  # you only need half of the fft list (real signal symmetry)
absList = [abs(elem) for elem in modifiedOriginal]
plt.plot(absList[:(d-1)],'b') 
# plt.ylim(0, 1.5)
# plt.xlim(0, 600000)
plt.show()
```

```python
# Autocorrelation 
standard = np.absolute(np.correlate(absList, absList, 'valid'))[0]

print(standard)
```

```python
fs, data = wav.read('original2.wav')

S, freqs, bins, im = plt.specgram(data.T[0], NFFT=1024, Fs=fs, noverlap=512)
# Plot a spectrogram
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
fs1, data1 = wav.read('Sample1.wav')

S1, freqs1, bins1, im1 = plt.specgram(data1.T[0], NFFT=1024, Fs=fs1, noverlap=512)
# Plot a spectrogram
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()

print(S1.shape)

C = np.cov(S, S1)

U, s, V = np.linalg.svd(C, full_matrices=False)
U.shape, V.shape, s.shape
print(max(s))
```

```python
fs2, data2 = wav.read('Sample2.wav')

S2, freqs2, bins2, im2 = plt.specgram(data2.T[0], NFFT=1024, Fs=fs2, noverlap=512)
# Plot a spectrogram
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()

print(S.shape)
print(S2.shape)

value1 = S.shape[1]
value2 = S2.shape[1]
print(value1)
print(value2)
minimum = min(value1, value2)
C2 = np.cov(S[:, :minimum], S2[:, :minimum])

U, s, V = np.linalg.svd(C2, full_matrices=False)
U.shape, V.shape, s.shape
print(max(s))
```
